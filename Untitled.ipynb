# In product master DF.

```python
def replace_pattern(string):
    pattern = r'(\d+)\^S'
    
    def replace_digit(digit_match):
        digit = int(digit_match.group(1))
        return f'Strip of {digit}'
    
    modified_string = re.sub(pattern, replace_digit, string)
    
    return modified_string
```

```python
variable_2 = df_2['Product_Desc']
split_words = ["CAPSULES", "TABLETS", "GEL","SYRUP","POWDER", "OINTMENT","INJECTION","DROP","CREAM", "SOAP", "VACCINE"]
my_list1 = []
for i in variable_2:
    match_counter = 0
    if "^S" in i:
        i = replace_pattern(i)
    #if i.find("CAPSULES", "TABLETS") != -1:
    for x in split_words:
        if x in i:
            i = i.split(x)
            j =  x + i[-1]
            k = [i[0], j]
            my_list1.append(k)
            match_counter = 1
    if match_counter == 0:
        k = [i]
        my_list1.append(k)
```

```python
df_22 = pd.DataFrame(df_2["Product_Desc"])

list_p_1 = []
list_p_2 = []

for elements in my_list1:
    if len(elements) == 2:
        list_p_1.append(elements[0])
        list_p_2.append(elements[-1])
    else:
        list_p_1.append(elements[0])
        list_p_2.append(np.nan)

df_22["First_part"] = list_p_1
df_22["Type"] = list_p_2
df_22["ED_MDM_Product_Code"] = df_2["ED_MDM_Product_Code"]
```

```python
new_df_3 = pd.read_csv("import.csv")
```

```python
new_df_3['Type_2'] = new_df_3['Type_2'].str.upper()
```

```python
from difflib import SequenceMatcher

# Assuming you already have the dataframe 'new_df_3'

# Function to calculate similarity score using SequenceMatcher
def similarity_score(a, b):
    return SequenceMatcher(None, str(a), str(b)).ratio()

# Create a new dataframe 'new_df_4' with the same columns as 'new_df_3'
new_df_4 = pd.DataFrame(columns=new_df_3.columns)

# Iterate over each row in 'new_df_3'
for index, row in new_df_3.iterrows():
    # Get the values from 'type' and 'type_2' columns in the current row
    type_value = row['Type']
    type_2_value = row['Type_2']
    
    # Calculate the similarity score
    score = similarity_score(type_value, type_2_value)
    
    # Check if the similarity score is more than 0.5 (50%)
    if score > 0.5:
        # Add the row to 'new_df_4'
        new_df_4 = new_df_4.append(row)

# Reset the index of 'new_df_4'
new_df_4 = new_df_4.reset_index(drop=True)
```
